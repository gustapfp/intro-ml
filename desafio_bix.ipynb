```python
from sklearn.svm import SVC, LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np


dataset2020 = pd.read_csv('base_vigente_2020.csv')
dataset_anos_anteriores = pd.read_csv('base_vigente_anos_anteriores.csv', sep=';')

mapa_dataset = {
    'class':'ventilacao'
}

ds_values = {
    'na':0
}
dataset2020.replace(ds_values, inplace=True)
dataset2020 = dataset2020.rename(columns = mapa_dataset)
dataset2020 = dataset2020.replace(['neg', 'pos'], [False, True])
dataset2020
```

```python
print(dataset2020.loc[:, dataset2020.columns != 'ventilacao'])
print(dataset2020.size)
```

```python
model = LinearSVC()
SEED = 5


x = dataset2020.loc[:, dataset2020.columns != 'ventilacao']
y = dataset2020[['ventilacao']]

np.random.seed(SEED)
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, stratify=y)

model.fit(x_treino, y_treino)
predictions = model.predict(x_teste)

acy_sre_linearSVC = accuracy_score(y_teste, predictions)
print(f"A precisão do modelo LinearSVC foi de: {(acy_sre_linearSVC)*100}")
```

```python
class CalculaCusto:
  def __init__(self, modelo, base_de_dados):
    self.modelo = modelo
    self.base_de_dados = base_de_dados

  def calcula_precisao(self):
    model = self.modelo
    SEED = 5
    np.random.seed(SEED)

    x = self.base_de_dados.loc[:, dataset2020.columns != 'ventilacao']
    y = self.base_de_dados[['ventilacao']]

    x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, stratify=y)

    model.fit(x_treino, y_treino)
    prediction = model.predict(x_teste)

    
    precisao = accuracy_score(y_teste, predictions)
    
    return (precisao*100), prediction

  def calcula_custo(self):
    precisao, prediction = self.calcula_precisao()
    taxa_de_erro = 100 - precisao

    tamanho = prediction.size # numero de diagnosticos
    numeros_de_true = np.sum(prediction) # numero de caminhões que tiveram problema no sistema de ventilação (pos)
    numeros_de_false = tamanho - numeros_de_true # nuemro de caminhões que não tiveram problema no sistema de ventilação (neg)

    numero_acerto = int(tamanho*(precisao/100)) # numero de previsões corretas segundo a precisão do modelo
    numero_erro = int(tamanho*(taxa_de_erro/100)) # numero de previsões incorretas segundo a precisão do modelo

    porc_acerto_true = int(numeros_de_true*(precisao/100)) # numero de diagnosticos corretos de problema no sistem de ventilaçaõ acertos
    porc_acerto_false = int(numeros_de_false*(precisao/100)) # numero de diagnosticos corretos de não haver problema no sistem de ventilaçaõ acertos com true

    porc_erros_true = int(numeros_de_true*(taxa_de_erro/100)) # numero de diagnosticos errados de problema no sistem de ventilaçaõ acertos
    porc_erros_false = int(numeros_de_false*(taxa_de_erro/100)) # numero de diagnosticos errados de  não haver problema no sistem de ventilaçaõ acertos
    
    gasto_acerto_true = porc_acerto_true*25 # Gasto com manutenção do sistema de ar com defeito $25
    gasto_acerto_false = porc_acerto_false*0 # Sem gasto com os caminhões com previsões de não ter problema no sistema pois não seriam mandados para a manutenção

    gasto_erros_true = porc_erros_true*500 # Gasto com manutenção corretiva $500
    gasto_erros_false = porc_erros_false*10 # Gasto com vistoria do caminhao pois nao tinham defeito no sistema
    gasto_total_modelo = gasto_acerto_true + gasto_acerto_false + gasto_erros_true + gasto_erros_false
    print(f'O modelo acertou {numero_acerto} diagnoósticos e errou {numero_erro}. Tendo uma precisão de:{precisao}%.')
    print(f'Com isso a IA conseguiu um custo de manutençaõo com os diagnosticos corretos de U${gasto_acerto_true}, com os diagnosticos errados os custos foram de U${gasto_erros_true + gasto_erros_false} totalizando um custo de U${gasto_total_modelo}')
    
```
